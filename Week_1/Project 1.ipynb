```python
import os
import cv2

# Dictionary containing UMC members' first names and surnames
umc_members = {
    "Enase": "Okonedo",
    "Chris": "Ogbechie",
    "Sola": "Oni",
    "Darlington": "Agholor",
    "Peter": "Bamokole"
}

# Path to the directory containing images
image_dir = r"C:\Users\CSC418\img"

def calculate_password_length(surname):
    return sum(1 for char in surname if char.isalpha())

def display_image(username):
    surname = umc_members.get(username)
    if surname:
        password = calculate_password_length(surname)
        entered_password = int(input("Enter your password (sum of alphabets in your surname): "))
        if entered_password == password:
            image_extensions = ['jpg', 'jpeg', 'png']
            image_found = False
            for ext in image_extensions:
                image_path = os.path.join(image_dir, f"{username}.{ext}")
                if os.path.exists(image_path):
                    image = cv2.imread(image_path)
                    cv2.imshow("UMC Member Image", image)
                    cv2.waitKey(0)
                    cv2.destroyAllWindows()
                    image_found = True
                    break
            if not image_found:
                print("Image not found.")
        else:
            print("Incorrect password.")
    else:
        print("Username not found.")

def main():
    username = input("Enter your username (first name): ")
    display_image(username)

if __name__ == "__main__":
    main()
```

```python
import cv2
import numpy as np
import matplotlib.pyplot as plt

# Function to apply separation (Splitting RGB channels)
def apply_separation(image):
    r, g, b = cv2.split(image)
    return r, g, b

# Function to apply addition (Adding Gaussian noise)
def apply_addition(image):
    noise = np.random.normal(0, 20, image.shape)
    noisy_image = image + noise
    return noisy_image

# Function to apply contrast enhancement
def apply_contrast(image):
    contrast = cv2.equalizeHist(image)
    return contrast

# Function to apply sharpening using Laplacian filter
def apply_sharpen(image):
    laplacian = cv2.Laplacian(image, cv2.CV_64F)
    sharpened = np.uint8(np.clip(image - 0.7 * laplacian, 0, 255))
    return sharpened

# Function to apply noise removal using Gaussian blur
def apply_noise_removal(image):
    denoised = cv2.GaussianBlur(image, (5, 5), 0)
    return denoised

# Function to apply scaling (Resizing)
def apply_scaling(image):
    scaled = cv2.resize(image, None, fx=0.5, fy=0.5)
    return scaled

# Function to display images
def display_images(original, enhanced, enhancement_name):
    plt.figure(figsize=(10, 6))
    plt.subplot(1, 2, 1)
    plt.title('Original Image')
    plt.imshow(cv2.cvtColor(original, cv2.COLOR_BGR2RGB))

    plt.subplot(1, 2, 2)
    plt.title(enhancement_name)
    plt.imshow(enhanced, cmap='gray')

    plt.tight_layout()
    plt.show()

# Function to perform image enhancement techniques
def image_enhancement(image_path, choice):
    # Load the image
    image = cv2.imread(image_path)
    gray = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)

    # Apply selected enhancement
    if choice == 1:
        enhanced = apply_separation(image)
        display_images(image, enhanced[0], 'Red Channel')
    elif choice == 2:
        enhanced = apply_addition(gray)
        display_images(image, enhanced, 'Noisy Image')
    elif choice == 3:
        enhanced = apply_contrast(gray)
        display_images(image, enhanced, 'Contrast Enhanced Image')
    elif choice == 4:
        enhanced = apply_sharpen(gray)
        display_images(image, enhanced, 'Sharpened Image')
    elif choice == 5:
        enhanced = apply_noise_removal(gray)
        display_images(image, enhanced, 'Noise Removed Image')
    elif choice == 6:
        enhanced = apply_scaling(gray)
        display_images(image, enhanced, 'Scaled Image')
    else:
        print("Invalid choice!")

# Dictionary mapping usernames to matnos and image paths
students = {
    'bruno': {'matno': '1111', 'image_path': 'C:\\Users\\ANDERSON AYE PEPPLE\\Documents\\andersonpeppleCSC418\\Week_2\\img\\bruno.jpg'},
    'david': {'matno': '2222', 'image_path': 'C:\\Users\\ANDERSON AYE PEPPLE\\Documents\\andersonpeppleCSC418\\Week_2\\img\\david.jpg'},
    'eze': {'matno': '3333', 'image_path': 'C:\\Users\\ANDERSON AYE PEPPLE\\Documents\\andersonpeppleCSC418\\Week_2\\img\\eze.jpg'},
    'francis': {'matno': '4444', 'image_path': 'C:\\Users\\ANDERSON AYE PEPPLE\\Documents\\andersonpeppleCSC418\\Week_2\\img\\francis.jpg'},
    'kaobi': {'matno': '5555', 'image_path': 'C:\\Users\\ANDERSON AYE PEPPLE\\Documents\\andersonpeppleCSC418\\Week_2\\img\\kaobi.jpg'},
    'lala': {'matno': '6666', 'image_path': 'C:\\Users\\ANDERSON AYE PEPPLE\\Documents\\andersonpeppleCSC418\\Week_2\\img\\lala.jpg'},
    'lotanna': {'matno': '7777', 'image_path': 'C:\\Users\\ANDERSON AYE PEPPLE\\Documents\\andersonpeppleCSC418\\Week_2\\img\\lotanna.jpg'},
    'oyinda': {'matno': '8888', 'image_path': 'C:\\Users\\ANDERSON AYE PEPPLE\\Documents\\andersonpeppleCSC418\\Week_2\\img\\oyinda.jpg'},
    'pese': {'matno': '9999', 'image_path': 'C:\\Users\\ANDERSON AYE PEPPLE\\Documents\\andersonpeppleCSC418\\Week_2\\img\\pese.jpg'},
    'zion': {'matno': '1010', 'image_path': 'C:\\Users\\ANDERSON AYE PEPPLE\\Documents\\andersonpeppleCSC418\\Week_2\\img\\zion.jpg'}
}

# Login
username = input("Enter username: ")
matno = input("Enter matno: ")

# Authentication
if username in students and students[username]['matno'] == matno:
    print("Login successful!")
    image_path = students[username]['image_path']
    print("Available enhancement techniques:")
    print("1. Separation (Splitting RGB channels)")
    print("2. Addition (Adding Gaussian noise)")
    print("3. Contrast Enhancement")
    print("4. Sharpening")
    print("5. Noise Removal")
    print("6. Scaling (Resizing)")
    choice = int(input("Enter your choice: "))
    image_enhancement(image_path, choice)
else:
    print("Login failed. Invalid username or matno.")
```
