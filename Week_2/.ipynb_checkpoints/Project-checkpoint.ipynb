```python
import cv2
import numpy as np

# Function to perform image enhancement techniques
def image_enhancement(image_path):
    # Load the image
    image = cv2.imread(image_path)

    # Convert image to grayscale
    gray = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)

    # Apply histogram equalization
    equalized = cv2.equalizeHist(gray)

    # Apply bilateral filter for smoothing while preserving edges
    smoothed = cv2.bilateralFilter(equalized, 9, 75, 75)

    # Apply sharpening using Laplacian filter
    laplacian = cv2.Laplacian(smoothed, cv2.CV_64F)
    sharpened = np.uint8(np.clip(equalized - 0.7 * laplacian, 0, 255))

    # Display the enhanced image
    cv2.imshow('Enhanced Image', sharpened)
    cv2.waitKey(0)
    cv2.destroyAllWindows()

# Dictionary mapping usernames to matnos and image paths
students = {
    'bruno': {'matno': '1111', 'image_path': 'C:\\Users\\ANDERSON AYE PEPPLE\\Documents\\andersonpeppleCSC418\\Week_2\\img\\bruno.jpg'},
    'david': {'matno': '2222', 'image_path': 'C:\\Users\\ANDERSON AYE PEPPLE\\Documents\\andersonpeppleCSC418\\Week_2\\img\\david.jpg'},
    'eze': {'matno': '3333', 'image_path': 'C:\\Users\\ANDERSON AYE PEPPLE\\Documents\\andersonpeppleCSC418\\Week_2\\img\\eze.jpg'},
    'francis': {'matno': '4444', 'image_path': 'C:\\Users\\ANDERSON AYE PEPPLE\\Documents\\andersonpeppleCSC418\\Week_2\\img\\francis.jpg'},
    'kaobi': {'matno': '5555', 'image_path': 'C:\\Users\\ANDERSON AYE PEPPLE\\Documents\\andersonpeppleCSC418\\Week_2\\img\\kaobi.jpg'},
    'lala': {'matno': '6666', 'image_path': 'C:\\Users\\ANDERSON AYE PEPPLE\\Documents\\andersonpeppleCSC418\\Week_2\\img\\lala.jpg'},
    'lotanna': {'matno': '7777', 'image_path': 'C:\\Users\\ANDERSON AYE PEPPLE\\Documents\\andersonpeppleCSC418\\Week_2\\img\\lotanna.jpg'},
    'oyinda': {'matno': '8888', 'image_path': 'C:\\Users\\ANDERSON AYE PEPPLE\\Documents\\andersonpeppleCSC418\\Week_2\\img\\oyinda.jpg'},
    'pese': {'matno': '9999', 'image_path': 'C:\\Users\\ANDERSON AYE PEPPLE\\Documents\\andersonpeppleCSC418\\Week_2\\img\\pese.jpg'},
    'zion': {'matno': '1010', 'image_path': 'C:\\Users\\ANDERSON AYE PEPPLE\\Documents\\andersonpeppleCSC418\\Week_2\\img\\zion.jpg'}
}

# Login
username = input("Enter username: ")
matno = input("Enter matno: ")

# Authentication
if username in students and students[username]['matno'] == matno:
    print("Login successful!")
    image_path = students[username]['image_path']
    print("Available enhancement techniques:")
    print("1. Image Enhancement")
    choice = int(input("Enter your choice: "))

    if choice == 1:
        image_enhancement(image_path)
    else:
        print("Invalid choice!")
else:
    print("Login failed. Invalid username or matno.")
```

```python
import cv2
import matplotlib.pyplot as plt
import numpy as np

def main():
    user_images = { 'bruno': (1, r'C:\Users\ANDERSON AYE PEPPLE\Documents\andersonpeppleCSC418\Week_2\img'), 
     'david': (2, r'C:\Users\hp\Documents\CSC418-asikabruno\week 2\img\david.jpg'), 
     'eze': (3, r'C:\Users\hp\Documents\CSC418-asikabruno\week 2\img\eze.jpg'), 
     'francis': (4, r'C:\Users\hp\Documents\CSC418-asikabruno\week 2\img\francis.jpg'), 
     'kaobi': (5, r'C:\Users\hp\Documents\CSC418-asikabruno\week 2\img\kaobi.jpg'), 
     'lala': (6, r'C:\Users\hp\Documents\CSC418-asikabruno\week 2\img\lala.jpg'),
     'lotanna': (7, r'C:\Users\hp\Documents\CSC418-asikabruno\week 2\img\lotanna.jpg'), 
     'oyinda': (8, r'C:\Users\hp\Documents\CSC418-asikabruno\week 2\img\oyinda.jpg'), 
     'pese': (9, r'C:\Users\hp\Documents\CSC418-asikabruno\week 2\img\pese.jpg'), 
     'zion': (10, r'C:\Users\hp\Documents\CSC418-asikabruno\week 2\img\zion.jpg') } 
    
    username = input("Enter username (first name): ") 
    password = int(input("Enter password (matric number): ")) 
    
    if username in user_images: 
        matric, image_path = user_images[username]
        expected_password = matric 
        if password == expected_password: 
            try:
                Option = input("Select an enhancement you would like to see. (Contrast, Sharpeness, Noise, Scaling, Transformation):") 
                if Option == "Contrast": 
                    image = cv2.imread(image_path)
                    plt.subplot(1,2,1)
                    plt.title(f'original {username}')
                    plt.imshow(image)
                    brightness = 5
                    contrast = 1.5
                    image2 = cv2.addWeighted(image, contrast, np.zeros(image.shape,image.dtype), 0, brightness)
                    cv2.imwrite(f'img/{username}_contrast.jpg', image2)
                    plt.subplot(1,2,2)
                    plt.imshow(image2)
                    plt.show() 
                
                elif Option == "Sharpeness":
                    image = cv2.imread(image_path)
                    plt.subplot(1,2,1)
                    plt.title(f'original {username}')
                    plt.imshow(image)
                    kernel = np.array([[0,-1,0],[-1,5,-1],[0,-1,0]])
                    sharpened_image = cv2.filter2D(image, -1, kernel)
                    cv2.imwrite(f'img/{username}_Sharpened.jpg', sharpened_image)
                    plt.subplot(1,2,2)
                    plt.imshow(sharpened_image)
                    plt.show()
                        
                elif Option == "Noise":
                    image = cv2.imread(image_path)
                    plt.subplot(1,2,1)
                    plt.title(f'original {username}')
                    plt.imshow(image)
                    filtered_image = cv2.medianBlur(image, 15)
                    cv2.imwrite(f'img/{username}_Noiseless.jpg', filtered_image)
                    plt.subplot(1,2,2)
                    plt.imshow(filtered_image)
                    plt.show()

                elif Option == "Scaling":
                    image = cv2.imread(image_path)
                    plt.subplot(1,2,1)
                    plt.title(f'original {username}')
                    plt.imshow(image)
                    scaled_image = cv2.resize(image, None, fx=2, fy=2)
                    cv2.imwrite(f'img/{username}_Scaled.jpg', scaled_image)
                    plt.subplot(1,2,2)
                    plt.imshow(scaled_image)
                    plt.show()

                elif Option == "Transformation":
                    image = cv2.imread(image_path)
                    plt.subplot(1,2,1)
                    plt.title(f'original {username}')
                    plt.imshow(image)
                    inverse_image = 255 - image
                    cv2.imwrite(f'img/{username}_Transformed.jpg', inverse_image)
                    plt.subplot(1,2,2)
                    plt.imshow(inverse_image)
                    plt.show()

                else:
                    print("Invalid Selection")
            except Exception as e: 
                print('Error:', e)

        else:
            print("Password Not seen")

    else:
        print("Authentication failed. User not found.")
        
main()
```
